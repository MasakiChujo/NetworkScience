## 中心性

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

### せっかくだし有名なネットワークを使おう
Zachary’s Karate Club graph.  
空手クラブのネットワーク

```python
G=nx.karate_club_graph()
nx.draw_spring(G)
```

### 次数中心性
ノード$N$個のネットワークを考え、ノード$i$の次数を$k_i$とすると
$$
\mathrm{DegreeCentrality}_i=\frac{k_i}{N-1}
$$

```python
deg=nx.degree_centrality(G)
vec=list(deg.values())
for i in range(len(vec)):
    vec[i]=vec[i]*3000
nx.draw_spring(G,node_size=vec,alpha=0.75)
```

### 媒介中心性
$$ 
c_B(v) =\sum_{s,t \in V} \frac{\sigma(s, t|v)}{\sigma(s, t)}
$$

```python
deg=nx.betweenness_centrality(G)
vec=list(deg.values())
for i in range(len(vec)):
    vec[i]=vec[i]*3000+10
nx.draw_spring(G,node_size=vec,alpha=0.75)
```

### 近接中心性
$$
C(u) = \frac{n - 1}{\sum_{v=1}^{n-1} d(v, u)}
$$

```python
deg=nx.closeness_centrality(G)
vec=list(deg.values())
for i in range(len(vec)):
    vec[i]=vec[i]*1000
nx.draw_spring(G,node_size=vec,alpha=0.75)
```
